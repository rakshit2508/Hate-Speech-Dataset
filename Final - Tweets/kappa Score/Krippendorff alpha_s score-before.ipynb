```python
import krippendorff
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_excel('tweets_list_modi_jan_5.xlsx')

df
```

```python
df.columns
```

```python
df1 = df[['Rakshit',
       'Ayan ', 'Shreyas']]
df1
```

```python
value_count = df1.to_numpy()
```

```python
print(value_count) 
```

```python
b=np.transpose(value_count) 
```

```python
b
```

```python

print("Krippendorff's alpha for nominal metric: ", krippendorff.alpha(reliability_data=b,
                                                                      level_of_measurement="nominal"))
print("Krippendorff's alpha for interval metric: ", krippendorff.alpha(reliability_data=b))
```

```python

print("Krippendorff's alpha for ordinal metric: ", krippendorff.alpha(reliability_data=b,
                                                                      level_of_measurement="ordinal"))
```
